```python
import numpy as np
import pandas as pd

np.random.seed(42)
N = 3000
```

```python
age = np.random.normal(30, 10, N).clip(10, 70)
systolic = np.random.normal(120, 20, N).clip(70, 180)
diastolic = (systolic * 0.6 + np.random.normal(0, 5, N)).clip(40, 120)
bs = np.random.normal(7, 1.8, N).clip(4, 20)
temp = np.random.normal(98.2, 0.7, N).clip(96, 103)
hr = np.random.normal(75, 12, N).clip(50, 120)
```

```python
risk_score = (
    (systolic > 140).astype(int) +
    (diastolic > 90).astype(int) +
    (bs > 10).astype(int) +
    (temp > 100).astype(int) +
    (hr > 90).astype(int)
)

risk = np.where(risk_score >= 3, "high risk",
        np.where(risk_score >= 1, "mid risk", "low risk"))
```

```python
df = pd.DataFrame({
    "Age": age.astype(int),
    "SystolicBP": systolic.astype(int),
    "DiastolicBP": diastolic.astype(int),
    "BS": np.round(bs, 2),
    "BodyTemp": np.round(temp, 1),
    "HeartRate": hr.astype(int),
    "RiskLevel": risk
})
```

```python
for col in ["Age", "SystolicBP", "DiastolicBP", "BS", "BodyTemp", "HeartRate"]:
    df.loc[df.sample(frac=0.10).index, col] = np.nan
```

```python
outlier_idx = df.sample(frac=0.03).index

df.loc[outlier_idx, "SystolicBP"] = np.random.randint(160, 180, len(outlier_idx))
df.loc[outlier_idx, "DiastolicBP"] = np.random.randint(100, 120, len(outlier_idx))
df.loc[outlier_idx, "BS"] = np.random.uniform(12, 20, len(outlier_idx))
df.loc[outlier_idx, "HeartRate"] = np.random.randint(100, 120, len(outlier_idx))
df.loc[outlier_idx, "BodyTemp"] = np.random.uniform(100, 103, len(outlier_idx))
```

```python
duplicates = df.sample(frac=0.01)
df = pd.concat([df, duplicates], ignore_index=True)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.to_csv("../data/synthetic_maternal_data.csv", index=False)
df.head()
```
